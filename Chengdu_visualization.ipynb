```python
from pandas import *
```

```python
table1 = pivot_table(gapminder,values = '口味',columns = ['类型'],aggfunc = np.mean)
print(table1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
table1.plot(kind = 'bar',rot = 0,title = 'Taste Score of Chuancai,Xiangcai in Chengdu')
plt.show()
print(table1)
```

```python
table2 = pivot_table(gapminder,values = '环境',columns= ['类型'],aggfunc = np.mean)
table2.plot(kind = 'bar',rot = 0,title = 'Environment Scores of Chuancai,Xiangcai in Chengdu')
plt.show()
print(table2)
```

```python
table3 = pivot_table(gapminder,values = '服务',columns= ['类型'],aggfunc = np.mean)
table3.plot(kind = 'bar',rot = 0,title = 'Service Scores of Chuancai,Xiangcai in Chengdu')
plt.show()
print(table3)
```

```python
csv_file = "成都商区分布.csv"
gapminder = pd.read_csv(csv_file)
form1 = gapminder[gapminder['类型']=="川菜"]
form2 = form1.sort_values(by='数量')
form_final = form2.tail(10)
print(form_final)
form_final.to_csv('form_final_川菜.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import *

chuancai = 'form_final_川菜.csv'
gapminder = pd.read_csv(chuancai)

table1 = pivot_table(gapminder,values = '数量',columns = ['地区'])
table1.plot(kind = 'bar',rot = 0,title = 'Chuan restaurant distrubution in Chengdu(top10)')
```
